```python
clean_df = df.dropna()
clean_df.tail()
```

```python
print(clean_df['Undergraduate Major'].loc[43])
print(clean_df['Mid-Career 90th Percentile Salary'].loc[43])
print(clean_df['Undergraduate Major'].loc[clean_df['Mid-Career 90th Percentile Salary'].idxmax()])
print(clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()])
clean_df['Starting Median Salary'].min()
	
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
low_risk = clean_df.sort_values('Spread',ascending=False)
print(low_risk)
```
